**Q.1.** Convert to Date Format (Convert-to-Date-Format.xlsx)

This time, you have been given a file containing Text dates into various formats. The challenge before
you is to write a single formula to convert them into an appropriate date format.

**Answer :-** Using Pandas

```python
import pandas as pd

# Read the file and specify which column is the date
customer_calls = pd.read_excel("Convert-to-Date-Format.xlsx")

# Output with dates converted to YYYY-MM-DD
customer_calls["Date"] = pd.to_datetime(customer_calls["Date"]).dt.strftime("%Y-%m-%d")
customer_calls.to_excel("Convert-to-Date-Format2.xlsx")
```
